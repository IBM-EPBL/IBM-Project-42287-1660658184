Removing outliers (creditScore, age)

```python
# for col in ['Age','CreditScore']:
#   mean = df[col].mean()
#   sd = df[col].std()

#   df = df[(df[col] <= mean + (3*sd))]

# df

#logic -> IQR {Inter Quartile Range}
# Q1 -> 1st quartile.(25%)
# Q2 -> 3rd quartile.(75%)
# Lowest range (lb) = Q1*1.5 
# Highest range (hb) = Q3*1.5
# Lies b/w them ok, else its a outlier
# if an outlier, replace with null value, change it as mean or 0 if required.


for x in ['Age', 'CreditScore']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan

df
```
